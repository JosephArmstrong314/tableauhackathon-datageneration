```python
import pandas as pd
```

```python
df = pd.read_csv('output.csv')
df2 = pd.read_csv('dummy_data.csv')

# just to clarify the geographic data is focused in the U.S.
df['country'] = ['USA' for _ in range(1000)]
```

```python
lst = [1,2,3,4]
for num in lst:
    df.drop(columns=[f'link{num}_converted'], inplace=True)
    df[f'link{num}_clicked'] = df[f'link{num}_clicked'].astype(int)
    
df['Page Visits'] = df2['# of Page Visits']
df['Conversion Rate'] = df2['Conversion Rate']
df['Click Through Rate'] = df2['Click Through Rate']
```

```python
cols = ['link1_clicked', 'link2_clicked', 'link3_clicked', 'link4_clicked']
df.loc[df[cols].eq(0).all(axis=1), ['Page Visits', 'Conversion Rate', 'Click Through Rate']] = 0
```

```python
data = df.to_csv('data.csv', index=False)
```
